# Шейкер

```python
#Смысл сортировки шейкером в проходе слева на право и справа на лево и сравнении элементов с целью нахождения максимального значения в диапазоне наших границ
test = [178,2019,56,-1020,12333,-78,-4,5,90,10] #Входные данные которые нужно рассортировать

left = 0 #Левая граница
right = len(test) - 1 #Правая граница с учетом сравнения текущего и следующего элемента в списке

while left <= right: # Условие при котором пока границы не пересекутся сортировка не остановится
    print("Слева на право:")
    for i in range(left, right): # Присваиваем переменной i индексы нашего списка
        #print(test)
        if test[i] > test[i + 1]: # Если при сравнении окажется что текущий элемент больше следующего
            test[i], test[i + 1] = test[i + 1], test[i] # Они меняются местами
    right -= 1 # Сужаем границу справа на 1 элемент т.к. мы узнали сортировкой самое большое число и его можно не сравнивать
    print(test)
    
    print("Справа на лево:")
    for i in range(right, left, -1): # Присваиваем переменной i индексы нашего списка справо на лево
        #print(test)
        if test[i - 1] > test[i]: # Если элемент левее больше нашего текущего
            test[i], test[i - 1] = test[i - 1], test[i] # Они меняются местами
    left += 1 # Сужаем левую границу на 1 т.к. мы нашли самый маленький элемент в итерации Справа на лево
    print(test)

print("Итог:",test) # Выводим результат

# Линейный поиск
# Идея в пробегании через каждый элемент по индексу и сравнении с нашим запрашиваемым значением
def linear_search(listik, n, target): 
    for i in range(0, n): # Индексы
        if (listik[i] == target): # Если находим элемент
            print("Ваше число под индексом: ",i)
            return i # Возвращаем индекс
    print("Вашего числа нету в списке :(")
    return -1 # Не находим элемент возвращаем -1

linear_search(test,len(test),int(input("Введите число, которое хотели бы найти ")))

```

# Сортировка вставками

```python
# Смысл сортировки вставками заключается в том, что начиная со второго элемента мы сравниваем выбранное число из списка со всеми
# предыдущими и вставляем его после элемента который меньше его не затрагивая остальные числа
test = [178,2019,56,-1020,12333,-78,-4,5,90,10]
print(test)
for i in range(1,len(test)):# Индексы от второго элемента
    print('Итерация: ',i)
    elem = test[i] # Элемент который будем сравнивать
    j = i-1 # Индекс предыдущего элемента
    while j>=0 and elem<test[j]: # Пока итерируемый индекс j предыдущих элементов >=0(для того чтобы в случае когда все элементы > того который сравниваем(elem), при j=-1 сравниваемый элемент становился в начало как самый меньший) и сравниваемый elem < чем элемент под индексом j
        j -= 1                   # уменьшаем индекс j для сравнения с элементом левее
    test.insert(j+1,test.pop(i)) # И когда j=-1(сравниваемый элемент вставляем в начало) или элемент под j всё таки меньше сравниваемого - вставляем elem на индекс j+1, при этом удаляя элемент функцией .pop
    #list.pop(index) - удаляет элемент в списке под индексом и возвращает значение элемента
    #list.insert(index,element) - вставляет элемент на указанный индекс раздвигая остальные элементы если это нужно
    print(test)
    
#Бинарный поиск
#Идея в поиске числа методом исключения спрашивая в диапазоне < оно среднего или > меняя при этом границы пока элемент не стал средним

def binary_search(listik, elem):
    low = 0 #Нижняя граница
    high = len(listik) - 1 #Верхняя граница
 
    while low <= high: # Пока нижняя и верхняя границы не зашли друг за друга
      
        middle = (low + high)//2 # Вводим среднее по отношению к заданным границам
       
        if listik[middle] == elem: # Если элемент равен среднему
            return middle # Выводим его индекс
        elif listik[middle] > elem: # Если больше среднего
            high = middle - 1 # Сужаем правую границу
        else:                 # Если меньше среднего
            low = middle + 1  # Сужаем левую

result = binary_search(test, int(input("Введите число, которое хотели бы найти ")))
if result: # Если индекс(result) не None
    print("Ваше число под индексом: ", str(result))
else:
    print("Нету такого числа (▀̿Ĺ̯▀̿ ̿)")
```

```python
# Пирамидальная сортировка заключается в представлении нашего списка в виде бинарного дерева в исходной последовательности:
#        0
#       | \
#      3   -1
#     | \  | \
#    4  10 3 -8
# Далее сравниваем выбранный корень и его два подкорня, начиная снизу и если подкорни больше чем корень меняем их местами
# Таким образом находим самый большой элемент который будет стоять в начале массива и будет являтся корнем всего дерева
# и меняем наш максимум с последним элементом списка, и из раза в раз находя больший элемент и дальше сортируя дерево
# преобразовываем наш список в сортированный

# Сортировка дерева, преобразовывающее его в двоичную кучу
def binary_heap_sort(arr, n, i):
    #Индексы:
    largest = i # Largest - корень нашей рассматриваемой части дерева
    left = 2 * i + 1   # Левый подкорень
    right = 2 * i + 2   # Правый подкорень

  # Проверяем существует ли левый дочерний элемент больший, чем корень
    if left < n and arr[i] < arr[left]:
        largest = left

  # Проверяем существует ли правый дочерний элемент больший, чем корень
    if right < n and arr[largest] < arr[right]:
        largest = right

    # Заменяем корень, если предыдущие условия могли выполнится и заменили largest на right или left
    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i] # Меняем корень с подкорнем
        print(arr)

        # Применяем heapify к корню.
        binary_heap_sort(arr, n, largest)
        print(arr)

# Основная функция для сортировки массива с использованием функции формирования двоичной кучи
def piramid_sort(arr):
    n = len(arr) # Количество элементов в массиве
    print("Построение двоичной кучи:")
    # Построение двоичной кучи, после которой совершается сортировка
    for i in range(n-1, -1, -1): # С самого правого по самый левый элементы с учетом
        print('Итерация: ',i)
        binary_heap_sort(arr, n, i) # Сортируем наше дерево для получения двоичной кучи
        print(arr)
    print(arr)
    print("_________")
    print("Начало сортировки:")
    # Сортировка элементов
    for i in range(n-1, 0, -1): #Берём индексы c самого правого до самого левого
        print('Итерация: ',i)
        arr[i], arr[0] = arr[0], arr[i] # Меняем первый элемент(Самый больший после функции сортировки) на последний
        print(arr)
        binary_heap_sort(arr, i, 0) # Снова сортируем элементы
        print(arr)

# Управляющий код для тестирования
arr = [12, 11, 13, 5, 6, 7,25,-128,-7,-68,56,1001,1,2,56,34,78,901,-1023,321312]
print(arr)
piramid_sort(arr)
print ("Итог: ", arr)

# Интерполяционный поиск 
# Это еще один алгоритм разделяй и властвуй но он начинает не с среднего значения относительно индексов а по формуле с предполагаемого индекса
# формула: index = low + ((target-listik[low])*(high-low) / (listik[high]-listik[low]))

def interpolation_search(listik, target):
    low = 0
    high = (len(listik) - 1)
    while low <= high and target >= listik[low] and target <= listik[high]: # Пока нижняя и верхняя границы не перешли через друг друга и искомое в наших границах
        index = low + int((target-listik[low])*(high-low) / (listik[high]-listik[low])) # Исполняем формулу поиска примерного индекса
        if listik[index] == target: # Если попали
            return index # Выводим
        if listik[index] < target: # Если меньше
            low = index + 1; # Сужаем нижнюю границу
        else:                # Если больше
            high = index - 1; # Сужаем верхнюю
    return -1

result = interpolation_search(arr, int(input("Введите число, которое хотели бы найти ")))
if result!=-1: # Если результат не -1(не нашло элемент)
    print("Ваше число под индексом: ", str(result))
else:
    print("Нету такого числа (▀̿Ĺ̯▀̿ ̿)")
```

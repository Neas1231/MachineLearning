```python
import random

player_stats = {"attack": 2,"defense":0,"hp":40,"max_hp":40,"xp":0,"max_xp":50,"level":0}

inventory = []  
    
mosters_lake = {}
    
monsters_castle = {}
    
    
    
class color:
   PURPLE = '\033[95m'
   CYAN = '\033[96m'
   DARKCYAN = '\033[36m'
   BLUE = '\033[94m'
   GREEN = '\033[92m'
   YELLOW = '\033[93m'
   RED = '\033[91m'
   BOLD = '\033[1m'
   END = '\033[0m'


def act_attack(atk):
    krit = 1
    if random.random() < 0.2:
        krit = 2
    damage = [atk*krit] # меньше шанс срабатывания крита
    return int(damage[0])
    
    

#def defence(defend)
#    if act_attack()


    
    
def start():
    print(color.BOLD + r'Добро пожаловать, начать игру?(y\n)' + color.END)
    agreement = input().lower()
    if agreement == 'y':
        print(color.BOLD + "Ваши начальные статистики:"+ color.END +f"\n Здоровье: {player_stats['hp']}, \
        Атака: {player_stats['attack']}, \
        Защита: {player_stats['defense']},\
        Опыт: {player_stats['xp']}")
        hub()
    elif agreement == "n":
        print('До встречи!')
        
    else:
        print(r'Я не понял вашей команды, повторите ещё раз!(y\n)')
        start()

def hub():
    print(color.BOLD + "Введите что вы хотите делать:" + color.END + "\n-Сражение\n-Инвентарь\n-Отдых\nОбратно")
    user_choose = input().lower()
    if user_choose == "сражение":
        choose_locations()
        
    elif user_choose == "инвентарь":
        inventory
    elif user_choose == "отдых":
        pass
    else:
        hub()

    
def choose_locations():
    print(color.BOLD + 'На какую локацию вы хотите войти?'+ color.END +'\n-Лес\nОбратно')
    user_location = input().lower()
    if user_location == "лес":
        forest()
    elif user_choose == "обратно":
        hub()
    else:
        choose_locations()


    
def forest():
    
    while True:
        monsters_forest = {"Мудрый дуб":{"attack": 4,"defense":1,"hp":10,"drop":{"Кора дуба":{"plus_hp":0,"plus_defense":1,"plus_attack":0}}},
                           "Крыса":{"attack": 3,"defense":0,"hp":15,"drop":{"Мышьяк":{"plus_hp":0,"plus_defense":1,"plus_attack":0}}}}
        enemy = random.choice(list(monsters_forest))
        print("\nНа тебя нападает "+color.RED+enemy+color.END+"!!")
        
        while True:
            if player_stats["hp"]<=0:
                raise Exception('Вы проиграли мухахахаха')
            user_battle_choose = input("-Атака\n-Защита\n-Инвентарь\n-Убежать\n\n").lower()
            
            
            if user_battle_choose == "атака":
                damage = act_attack(player_stats['attack'])
                monsters_forest[enemy]['hp'] -= damage
                if monsters_forest[enemy]['hp']>0:
                    print(color.RED + f"\nВаша атака нанесла {damage} урона! " + color.END + f"Осталось {monsters_forest[enemy]['hp']} здоровья у монстра")
                    player_stats['hp'] -= act_attack(monsters_forest[enemy]['attack'])
                    print(color.RED + f'Монстр нанёс {act_attack(monsters_forest[enemy]["attack"])} урона'+ color.END)
                    print(f'У вас осталось {player_stats["hp"]} здоровья')
                else:
                    print(f'Вы победили {enemy}!')
                    item = random.choice(list(monsters_forest[enemy]['drop']))
                    print(f'Полученный предмет:{item}\n')
                    inventory.append(item)
                    break
            elif user_battle_choose == "защита":
                print("Вы защищаетесь! 50% урона противника нивелируется")
                player_stats['hp'] -= act_attack(monsters_forest[enemy]['attack'])
                print(color.RED + f'Монстр нанёс {act_attack(monsters_forest[enemy]["attack"])} урона'+ color.END)
                print(f'У вас осталось {player_stats["hp"]} здоровья')
                
                
            elif user_battle_choose == "инвентарь":
                inventory()
                player_stats['hp'] -= act_attack(monsters_forest[enemy]['attack'])
                print(color.RED + f'Монстр нанёс {act_attack(monsters_forest[enemy]["attack"])} урона'+ color.END)
                print(f'У вас осталось {player_stats["hp"]} здоровья')
                
            
            elif user_battle_choose == "убежать":
                if random.random() < 0.4:
                    hub()
                else:
                    print("Вы не смогли убежать :(")
                    player_stats['hp'] -= act_attack(monsters_forest[enemy]['attack'])
                    print(color.RED + f'Монстр нанёс {act_attack(monsters_forest[enemy]["attack"])} урона'+ color.END)
                    print(f'У вас осталось {player_stats["hp"]} здоровья')
                    
            
            
            else:
                print("Вы неправильно ввели слово и за это пропускаете ход!!")
                player_stats['hp'] -= act_attack(monsters_forest[enemy]['attack'])
                print(color.RED + f'Монстр нанёс {act_attack(monsters_forest[enemy]["attack"])} урона'+ color.END)
                print(f'У вас осталось {player_stats["hp"]} здоровья')
    

start()
 
```

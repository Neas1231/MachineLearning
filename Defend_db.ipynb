```python
test = [1,2,3,'55e',4,5,'g',6,7,8,0,'6','g']
# функция убирающая не цифры
def func(test):
    test_result = []
    for num in map(str,test):
        if num.isdigit():
            test_result.append(int(num))
    print(test_result)
func(test)
```

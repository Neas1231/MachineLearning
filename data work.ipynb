```python
def up_low(ini_df):
    work_val = ini_df.copy()
    if work_val.dtypes!=np.object:
        Q1 = np.percentile(work_val, 25, interpolation = 'midpoint')
        Q3 = np.percentile(work_val, 75, interpolation = 'midpoint')
        IQR = Q3-Q1
        upper = Q3+1.5*IQR
        lower = Q1-1.5*IQR
    return upper,lower
```

```python
df_copy.select_dtypes(exclude=object).columns
```

```python
sns.boxplot(y='Latitude',data=df_copy)
plt.show()
upper,lower = up_low(df_copy['Latitude'])
df_copy = df_copy[~((df_copy['Latitude']>upper) | (df_copy['Latitude']<lower))] 
df_copy
sns.boxplot(y='Latitude',data=df_copy)
plt.show()
```

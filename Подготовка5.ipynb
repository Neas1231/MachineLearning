### PUBLICATION_NUMBER

```python
def pub_num_first(num):
    return num[:2]
    
def pub_num_second(num):
    return num[2+1:]
    
df_cop['Pub_num_seq'] = df['PUBLICATION_NUMBER'].apply(pub_num_first)
df_cop['Pub_num_year'] = df['PUBLICATION_NUMBER'].apply(pub_num_second)
```

```python
df_cop
```

### PUBLICATION_DATE

```python
def pub_date_day(date):
    return date[:2]
def pub_date_month(date):
    return date[2+1:5]
def pub_date_year(date):
    return date[5+1:]
df_cop["Pub_date_day"] = df['PUBLICATION_DATE'].apply(pub_date_day)
df_cop["Pub_date_month"] = df['PUBLICATION_DATE'].apply(pub_date_month)
df_cop["Pub_date_year"] = df['PUBLICATION_DATE'].apply(pub_date_year)
```

```python
df_cop
```

### DATE_OF_FILING

```python
def date_fill(date):
    indx = [i for i, ltr in enumerate(date) if ltr == '/']
    if indx:
        return [date[:indx[0]],date[indx[0]+1:indx[1]],date[indx[1]+1:]]
s = df['DATE_OF_FILING'].apply(date_fill)
df_date_fill = pd.DataFrame(dict(zip(s.index, s.values))).T
df_date_fill.columns = ['Date_fill_day','Date_fill_month','Date_fill_year']
df_date_fill
```

```python
df_cop = pd.concat([df_cop, df_date_fill], axis=1)
df_cop
```

### Обработка текстовой информации

Приведём текстовые данные к форме инфинитивов и удалим всевозможные ненужные символы

```python
import re
from pymorphy2 import MorphAnalyzer
from nltk.corpus import stopwords
patterns = "[0-9!#$%&'()*+,./:;<=>?@[\]^_`{|}~—\"\-]+"
stopwords_ru = stopwords.words("english")
morph = MorphAnalyzer()
def lemmatize(doc):
    doc = re.sub(patterns, ' ', doc)
    tokens = []
    for token in doc.split():
        if token and token not in stopwords_ru:
            token = token.strip()
            token = morph.normal_forms(token)[0]
            
            tokens.append(token)
    #if len(tokens) > 2:
    return tokens
    #return None
```

```python
df_cop['Title_of_invention'] = df_cop['Title_of_invention'].apply(lemmatize)
df_cop['Title_of_invention'] = df_cop['Title_of_invention'].apply(list)
df_cop
```

```python
def adress_fuller(row):
    row[['district','street','underground']] = row[['district','street','underground']].fillna(row['city'])
    return row

df.loc[df['city'] != 'moscow'] = df.loc[df['city'] != 'moscow'].apply(adress_fuller,axis=1)
df
```

```python
df.loc[df['city'] != 'moscow']
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns=['price'])
X.loc[:,['city','district','street','underground','residential_complex']] = X.loc[:,['city','district','street','underground','residential_complex']].fillna('No_data')
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_test.isna().sum()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
from catboost import CatBoostRegressor
model = CatBoostRegressor(nan_mode='Max',cat_features=['city','district','street','underground','residential_complex'],learning_rate=0.3, depth=7,
                          loss_function='MAE',l2_leaf_reg = 4,
                          iterations = 1200,leaf_estimation_method ="Exact",
                          verbose = 10,eval_metric='MAE'
)
model.fit(X_train.drop(columns = 'price_per_m2'), y_train)
y_pred = model.predict(X_test.drop(columns = 'price_per_m2'))
```

```python
y_pred
```

```python
print(mean_absolute_error(y_test,y_pred))
```

```python
y_pred_perm2 = y_pred/X_test['total_meters']
print('mean_absolute_error:' , mean_absolute_error(X_test['price_per_m2'],y_pred_perm2))
print('mean_squared_error:' , mean_squared_error(X_test['price_per_m2'],y_pred_perm2))
print('r2_score:' , r2_score(X_test['price_per_m2'],y_pred_perm2))
print('mean_absolute_percentage_error:' , mean_absolute_percentage_error(X_test['price_per_m2'],y_pred_perm2))
```

```python
df_cop = df.copy()
```

```python
df_cop.drop(columns=['residential_complex'], inplace=True)
df_cop.dropna(inplace=True)
def quantiles(arr):
    work_val = arr.copy()
    Q1 = work_val.quantile(0.25)#np.percentile(work_val, 25, method= 'midpoint')
    Q3 = work_val.quantile(0.75)#np.percentile(work_val, 75, method= 'midpoint')
    IQR = Q3-Q1
    upper = Q3+1.5*IQR
    lower = Q1-1.5*IQR
    return upper,lower

def anti_emmi(df):
    if df.select_dtypes(include = [np.number]).all().any():
        for col in df.select_dtypes(include = [np.number]).columns:

            up,low = quantiles(df[col])
            df = df[~(df[col] < low) & ~(df[col] > up)]

    return df.reset_index(drop=True)

df_cop = anti_emmi(df_cop)
df_cop
```

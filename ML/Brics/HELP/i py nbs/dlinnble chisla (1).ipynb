```python
# Граф являющийся словарём списков(т.к. нам нужно знать только с кем соиденена вершина)
graph = {
    1: [2,5,7],
    2: [1,3,4,6],
    3: [2,5,6],
    4: [2,8,9],
    5: [1,3],
    6: [2,3],
    7: [1],
    8: [4,9],
    9: [4,8],
}

visited = []  # Посещена ли вершина?
Query = []  # Посещаемая очередь
BFS = [] # Конечный путь посещений

# Поиск в ширину - ПВШ (Breadth First Search - BFS)
def bfs(node):
    if node in visited:  # Если вершина уже посещена, выходим
        return
    visited.append(node)  # Посетили вершину v
    BFS.append(node)  # Запоминаем порядок обхода
    for i in graph[node]:  # Все смежные с v вершины
        if not i in visited: # Если не посещена добавляем в очередь
            Query.append(i)
    while Query: # И пока есть очередь
        bfs(Query.pop(0)) # Мы продолжаем убирать значения из очереди и заполнять их по мере продвижения

start = 7
bfs(start)  # start - начальная вершина обхода
print(BFS)
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.tokenize import word_tokenize
from pymorphy3 import MorphAnalyzer
import re
patterns = "[A-Za-z0-9!#$%&'()*+,./:;<=>?@[\]^_`{|}~—\"\-]+"
stopwords_ru = stopwords.words("russian")
morph = MorphAnalyzer()
stemmer = PorterStemmer()
def lemmatize(text):
    doc = re.sub(patterns, ' ', text)
    tokens = word_tokenize(doc, language='russian')
    lemmas = [morph.parse(token)[0].normal_form for token in tokens if token.lower() not in stopwords_ru]
    stems = [stemmer.stem(token) for token in tokens]
    return stems


df_cop['comment'] = df_cop['comment'].apply(lemmatize)
```

```python
def concat(text):
    text= ' '.join(text)
    return text

df_cop['comment'] = df_cop['comment'].apply(concat)
```

```python
import pandas as pd
df_cop = pd.read_csv('tokenized_data.csv')
```

```python
df_cop.drop(columns='Unnamed: 0',inplace=True)
df_cop
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_counts = count_vect.fit_transform(df_cop['comment'])
X_counts.shape
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer().fit(X_counts)
X_tfidf = tf_transformer.transform(X_counts)
X_tfidf.shape
```

```python
from sklearn.decomposition import TruncatedSVD
clf = TruncatedSVD(30)
X_decom = clf.fit_transform(X_tfidf)
X_decom
```

```python
X = X_decom.copy()
y = df_cop['label']
```

```python
X = X_tfidf.copy()
y = df_cop['label']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from imblearn.under_sampling import RandomUnderSampler
undersampler = RandomUnderSampler()
X_train, y_train = undersampler.fit_resample(X_train, y_train)
```

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report

# Получение весов классов
#class_weights = compute_class_weight('balanced', classes=np.unique(y), y=y)

# Создание модели с взвешиванием классов
model = RandomForestClassifier()
model.fit(X_train,y_train)
print(classification_report(y_test,model.predict(X_test)))
```

```python
from imblearn.over_sampling import SMOTE 
sm = SMOTE(random_state=42)

X_train, y_train = sm.fit_resample(X_train, y_train)
```

```python
from imblearn.over_sampling import ADASYN
adasyn = ADASYN()

X_train, y_train = adasyn.fit_resample(X_train, y_train)
```

```python
from sklearn.naive_bayes import MultinomialNB, GaussianNB, BernoulliNB
from sklearn.metrics import classification_report

clf2 = GaussianNB()
clf3 = BernoulliNB()

clf2.fit(X_train,y_train)
clf3.fit(X_train,y_train)
print(classification_report(y_test,clf2.predict(X_test)),classification_report(y_test,clf3.predict(X_test)),sep = '\n')
```
